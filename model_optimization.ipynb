```python
import torch
from transformers import AutoModelForImageClassification
import os

# 1. Model Selection: โหลด Pre-trained ResNet-50 จาก Hugging Face
model_name = "microsoft/resnet-50"
model = AutoModelForImageClassification.from_pretrained(model_name)
model.eval() # ปรับเป็นโหมดใช้งานจริง

# บันทึกเป็น .pt เพื่อเอาไว้ดูขนาดไฟล์ Baseline
torch.save(model.state_dict(), "hf_resnet50.pt") # hf = hugging face

# 2. Create Dummy Input (1 image, 3 channels, 224x224)
# สำหรับโมเดล Vision ของ Hugging Face มักจะใช้ชื่อ input ว่า pixel_values
dummy_input = torch.randn(1, 3, 224, 224)
```

```python
# 3. Export to ONNX
torch.onnx.export(
    model,               # โมเดลต้นฉบับ (ResNet-50)
    dummy_input,         # ข้อมูลตัวอย่างสำหรับไล่กราฟ (Tracing)
    "hf_resnet50.onnx",     # ชื่อไฟล์ปลายทาง
    input_names=['input'],   # กำหนดชื่อเรียก Input
    output_names=['output'], # กำหนดชื่อเรียก Output
    opset_version=14     # เวอร์ชันของชุดคำสั่งคณิตศาสตร์
)
print("Exported to ONNX successfully!")
```

```python
from onnxruntime.quantization import quantize_dynamic, QuantType, quant_pre_process

# เตรียมไฟล์ก่อน Quantize (จัดระเบียบกราฟ)
quant_pre_process("hf_resnet50.onnx", "hf_resnet50_prep.onnx")

# 4. Quantization: บีบอัดไฟล์เป็น INT8
quantize_dynamic(
    "hf_resnet50_prep.onnx",              
    "hf_resnet50_int8.onnx",         
    weight_type=QuantType.QUInt8,   # แปลง Weights เป็น 8-bit Integer
)

print("Quantization Complete! Check your folder for the new file.")
```

```python
import time
import onnxruntime as ort
import pandas as pd

def benchmark_pytorch(model, input_tensor, runs=100):
    times = []
    with torch.no_grad():
        for _ in range(runs):
            start = time.time()
            _ = model(input_tensor) # ประมวลผลด้วย PyTorch
            end = time.time()
            times.append(end - start)
    return (sum(times) / runs) * 1000

def benchmark_onnx(model_path, input_tensor, runs=100):
    session = ort.InferenceSession(model_path)
    # ดึงชื่อ Input อัตโนมัติ (ซึ่งก็คือ pixel_values)
    input_name = session.get_inputs()[0].name
    ort_inputs = {input_name: input_tensor.numpy()}
    
    times = []
    for _ in range(runs):
        start = time.time()
        session.run(None, ort_inputs) # ประมวลผลด้วย ONNX
        end = time.time()
        times.append(end - start)
    return (sum(times) / runs) * 1000

def get_size_mb(file):
    return os.path.getsize(file) / (1024*1024)

# รันเก็บค่า Latency
print("Benchmarking PyTorch...")
pt_latency = benchmark_pytorch(model, dummy_input)

print("Benchmarking ONNX...")
onnx_latency = benchmark_onnx("hf_resnet50_prep.onnx", dummy_input)

print("Benchmarking ONNX INT8...")
int8_latency = benchmark_onnx("hf_resnet50_int8.onnx", dummy_input)

# 5. Data Collection
data = {
    "Model": ["HF PyTorch FP32", "HF ONNX FP32", "HF ONNX INT8"],
    "File Size (MB)": [
        get_size_mb("hf_resnet50.pt"),
        get_size_mb("hf_resnet50_prep.onnx"),
        get_size_mb("hf_resnet50_int8.onnx")
    ],
    "Avg Latency (ms)": [pt_latency, onnx_latency, int8_latency]
}

df = pd.DataFrame(data)
print("\n--- สรุปผล Model Optimization ---")
print(df)
```
